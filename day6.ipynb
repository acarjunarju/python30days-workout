```python
c = int(input("enter the number "))
if c<25:
  if c%2 ==0:
    print("c is even numberless than 25 ")
  else:
    print("c is odd number less than 25")
else:
  print("c is greater than 25")
```

If-elif -else statement

```python
var = input("enter the alphabet: ")
if var == "a":
  print("this is vowel")
elif var == "e":
  print("this is vowel")
elif var == "i":
  print("this is vowel")
elif var == "o":
  print("this is vowel")
elif var == "u":
  print("this is vowel")
else:
  print("this is consonant")
```
