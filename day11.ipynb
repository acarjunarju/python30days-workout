```python
class ShoppingCart:
    def __init__(self, items=None):
        if items is None:
            self.items = []
        else:
            self.items = items

    def add_item(self, item):
        self.items.append(item)

    def remove_item(self, item):
        if item in self.items:
            self.items.remove(item)
        else:
            print("Item not in cart")

    def view_cart(self):
        print("Items:", self.items)

c = ShoppingCart()
c.add_item("apple")
c.add_item("banana")
c.view_cart()
c.remove_item("apple")
c.view_cart()
```

```python
class Library:
    def __init__(self, book=None):
        if book is None:
            self.book = []
        else:
            self.book = book

    def add_book(self, book):
        self.book.append(book)

    def display_book(self):
        print("book:", self.book)

b = Library()
b.add_book("The Great Gatsby")
b.display_book()
```

```python
class Students:
  def __init__(self, name, marks=None):
    self.name = name
    self.marks = marks if marks is not None else []
  def std_name(self):
    print("name:", self.name)
  def marks_list(self):
    print("marks:", self.marks)
  def average_marks(self):
    if not self.marks:
        print("average marks: 0")
    else:
        print("average marks:", sum(self.marks) / len(self.marks))

# Properly instantiate the class
s = Students("arjun", [20, 23, 40, 40, 32, 44, 43, 34])
s.std_name()
s.marks_list()
s.average_marks()
```
